```python
import pandas as pd
import numpy as np
import networkx as nx 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("routes.csv")
df.head()
```

```python
g = nx.Graph()

g = nx.from_pandas_edgelist(df, source = 'Source Airport', target = 'Destination Airport')
```

```python
#g.nodes
#g.edges
```

```python
# Degree Centrality - no of direct connections
dc = nx.degree_centrality(g)  
```

```python
print(dc)
```

```python
sorted_dc = {k: v for k, v in sorted(dc.items(), key=lambda item: item[1], reverse=True)}
sorted_dc
```

```python
# closeness centrality
closeness = nx.closeness_centrality(g)
print(closeness)
```

```python
sorted_closeness = {k: v for k, v in sorted(closeness.items(), key=lambda item: item[1], reverse=True)}
sorted_closeness
```

```python
## Betweeness Centrality 
b = nx.betweenness_centrality(g) 
print(b)
```

```python
sorted_b = {k: v for k, v in sorted(b.items(), key=lambda item: item[1], reverse=True)}
sorted_b
```
