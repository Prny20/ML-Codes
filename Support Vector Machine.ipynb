```python
iimport numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.svm import LinearSVC
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("bank.csv")
```

```python
df.head()
```

```python
# feature and target
# train test split

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

## LinearSVC

```python
lin_svc = LinearSVC(random_state=1)
lin_svc.fit(X_train,y_train)
```

```python
y_pred = lin_svc.predict(X_test)
print(classification_report(y_test,y_pred))
```

## Soft margin

```python
lin_svc_soft = LinearSVC(random_state=1, C=0.5)
lin_svc_soft.fit(X_train,y_train)
```

```python
y_pred = lin_svc_soft.predict(X_test)
print(classification_report(y_test,y_pred))
```

## SVC polynomial

```python
svc = SVC(kernel="poly", degree=1, random_state=1)
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
print(classification_report(y_test,y_pred))
```

## SVC Radial Basis

```python
svc = SVC(kernel="rbf",random_state=1,C=0.5)
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
print(classification_report(y_test,y_pred))
```
